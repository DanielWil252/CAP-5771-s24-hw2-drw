```python
#function to find entropy.
import numpy as np
def entropy_impurity(labels):#generated this entropy function from chatgpt using the one I wrote for question 2 as a basis. Asked to write it in a manner that accepts labels.
    unique_labels, label_counts = np.unique(labels, return_counts=True)
    print(f"unique labels:{unique_labels} label_counts: {label_counts}")
    total_samples = len(labels)
    entropy = 0
    for count in label_counts:
        probability = count / total_samples
        entropy -= probability * np.log2(probability)
    return entropy

def infogain(entropies,probabilities,parentEntropy,parentTotal=1):
    infogain_val = parentEntropy
    for index,val in enumerate(entropies):
        weight = probabilities[index]/parentTotal
        infogain_val -= weight*val
    return infogain_val

def get_best_split(dataNames,data,totalSamples,parentEntropy):#finding the best split given features and sample number. If we split from any of the 4 classes, which gives us the most information gain for lung cancer?
    
    bestInfoGain = -float("inf")
    bestSplit = "all are the same."
    for index,name in enumerate(dataNames):
        print(f"testing for name {name}")
        tempRight = np.array([])
        tempLeft = np.array([])
        for row in range(np.size(data[:,-1])):
            if data[row,index] == False: #change this to flip the directions for the tree.
                if tempRight.size == 0:
                    tempRight = np.array(data[row,:])
                else:
                    tempRight = np.vstack((tempRight,data[row,:]))
            else:
                if tempLeft.size == 0:
                    tempLeft = np.array(data[row,:])
                else:
                    tempLeft = np.vstack((tempLeft,data[row,:]))
        #compute entropy for each.
        print(tempRight.size)
        print(tempLeft.size)
        #print(f"tempRight:\n {tempRight}\n tempLeft:\n {tempLeft}")
        if tempRight.size != 0 and tempLeft.size != 0:
            
            entropyRight = 0 if tempRight.ndim == 1 else entropy_impurity(np.reshape(tempRight[:,-1],np.size(tempRight[:,-1])))
            entropyLeft = 0 if tempLeft.ndim == 1 else entropy_impurity(np.reshape(tempLeft[:,-1],np.size(tempLeft[:,-1])))
            #compute probabilities.
            probabilityRight = 1/totalSamples if tempRight.ndim == 1 else np.size(tempRight[:,0])/totalSamples
            probabilityLeft = 1/totalSamples if tempLeft.ndim == 1 else np.size(tempLeft[:,0])/totalSamples
            print(f"probability right: {probabilityRight} probability left: {probabilityLeft}")
            #compute information gain.
            tempInfoGain = infogain([entropyLeft,entropyRight],[probabilityLeft,probabilityRight],parentEntropy,parentTotal=totalSamples)

            print(f"information gain: {tempInfoGain}")
            if(tempInfoGain > bestInfoGain):
                bestInfoGain = tempInfoGain
                bestSplit = name
                newLeft = tempLeft
                newRight = tempRight
        else:
            print(f"splitting name {name} will create an empty node.")
    print(f"Best Split: {bestSplit} information gain: {bestInfoGain}")
    return bestSplit,bestInfoGain, newLeft, newRight
```

```python
col_names = ["Tobacco Smoking","Radon Exposure","Chronic Cough","Weight Loss"]#excluding lung cancer itself.
data = [[True,True,True,False,True],
        [True,False,True,False,True],
        [True,False,True,True,True],
        [True,False,True,True,True],
        [False,True,False,True,True],
        [True,False,False,False,False],
        [False,False,True,False,False],
        [False,False,True,True,False],
        [False,False,True,True,False],
        [False,False,False,True,False]]
data = np.array(data)
rootEntropy = entropy_impurity(np.reshape(data[:,-1],10)) #2 possibilities with equal chance
print(rootEntropy)#should be one
bestFirstSplit, firstSplitInformationGain,newLeft,newRight = get_best_split(col_names,data,np.size(data[:,-1]),rootEntropy)

secondSplitLeftEntropy = entropy_impurity(np.reshape(newLeft[:,-1],np.size(newLeft[:,-1])))
bestSecondSplitLeft, secondSplitLeftInformationGain, secondSplitLeftnewLeft,secondSplitLeftnewRight = get_best_split(col_names,newLeft,np.size(newLeft[:,-1]),secondSplitLeftEntropy)
#note, the chronic cough split creates pure leaves.

secondSplitRightEntropy = entropy_impurity(np.reshape(newRight[:,-1],np.size(newRight[:,-1])))
bestSecondSplitRight, secondSplitRightInformationGain, secondSplitRightnewLeft,secondSplitRightNewRight = get_best_split(col_names,newRight,np.size(newRight[:,-1]),secondSplitRightEntropy)
#Radon Exposure split creates a pure node.

print(f"best first split: {bestFirstSplit} {firstSplitInformationGain} best second split left: {bestSecondSplitLeft} {secondSplitLeftInformationGain} best second split right: {bestSecondSplitRight} {secondSplitRightInformationGain}")
```

```python
def gini(probabilities, total=1):
    gini_val = 1
    for num in probabilities:
        if num !=0:
            num /= total
            gini_val += -num**2
    return gini_val



#10 instances of C0
#10 instances of C1
overallGini = gini([.5,.5])
print(overallGini)

#20 unique instances of customer ID
#however, since there is only one instance, the gini is 1-1^2, which is 0.

#10 instances of M
#4 are C1
#6 are C0
#10 instances of F
#6 are C1
#4 are C0
maleGini = gini([.6,.4])
femaleGini = gini([.4,.6])

genderGini = .5*(maleGini + femaleGini)
print(genderGini)

#3 types
#4 instances of family
#1 is C0
#3 are C1
#8 instances of sports
#8 are C0
#0 are C1
#8 instances of luxury
#1 is C0
#7 are C1
familyGini = gini([.25,.75])
sportsGini = gini([0,1])
luxuryGini = gini([.125,.875])

carGini = .2*familyGini + .4*sportsGini + .4*luxuryGini
print(carGini)

#4 types
#5 instances of small
#3 are C0
#2 are C1
#7 instances of medium
#3 are C0
#4 are C1
#4 instances of large
#2 are C0
#2 are C1
#4 instances of extra large
#2 are C0
#2 are C1
smallGini = gini([3,2],total=5)
mediumGini = gini([3,4],total=7)
largeGini = gini([2,2],total=4)
extraLargeGini = gini([2,2],total=4)

shirtSizeGini = .25*smallGini + .35*mediumGini + .2*largeGini + .2*extraLargeGini
print(shirtSizeGini)
```

```python
def splitInfo(probabilities,parentTotal=1):#adapted from infogain
    splitInfo_val = 0
    for nums in probabilities:
        proportion = nums/parentTotal
        if proportion > 0:
            splitInfo_val -= proportion*np.log2(proportion)
    return splitInfo_val



root_labels = np.array([1]*10 + [0]*10)  # positive = 1; negative = 0. Entropy is 1 since odds are 50/50

rootEntropy = entropy_impurity(root_labels)
print(rootEntropy)

#ID info gain is the same as the root entropy, since , since the entropy for the Ids is 0.

#Handedness:
#10 from left handed
#9 positive
#1 negative
#10 from right handed
#1 positive
#9 negative

left_handed_labels = np.array([1]*9 + [0]*1)

leftHandedEntropy = entropy_impurity(left_handed_labels)

right_handed_labels = np.array([1]*1 + [0]*9)

rightHandedEntropy = entropy_impurity(right_handed_labels)

handednessInfogain = infogain([leftHandedEntropy,rightHandedEntropy],[np.size(leftHandedEntropy),np.size(rightHandedEntropy)],rootEntropy,parentTotal=20)
print(handednessInfogain)

#finding gain ratio for both.
splitInfoID = splitInfo([1]*np.size(root_labels),parentTotal=np.size(root_labels))

gainRatioID = 1/splitInfoID
print(gainRatioID)

splitInfoHandedness = splitInfo([np.size(leftHandedEntropy),np.size(rightHandedEntropy)],parentTotal=20)
print(splitInfoHandedness)
```
